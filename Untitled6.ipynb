```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"D:\Code\Average.csv")
display(dataset)
```

```python
Year = dataset["Month"]
A1 =dataset["2019"]
A2 =dataset["2020"]
A3 = dataset["2021"]

rcy = plt.figure()
rcy.set_figwidth(10)
rcy.set_figheight(7)

barTry=0.11
X_axis=np.arange(len(Year))
plt.bar(X_axis -0.2,A1, color='orange',width=barTry,label='2019')
plt.bar(X_axis -0.1,A2, color='green',width=barTry,label='2020')
plt.bar(X_axis,A3, color='purple',width=barTry,label='2021')
plt.xticks(X_axis,Year)

plt.title("Monthly Average Farmgate Price of Chicken Egg (Commercial), Philippines: April-June 2019-2021", fontweight="bold")
plt.xlabel("Month")
plt.ylabel("Average Farmgate Price")

plt.grid(color = 'black', linewidth = 0.5)
plt.legend()
plt.show()
```
